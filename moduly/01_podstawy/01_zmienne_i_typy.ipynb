## 1. Czym jest zmienna?

Zmienna to nazwane "pudełko", w którym przechowujesz wartość. W Pythonie tworzysz ją po prostu pisząc:

```python
nazwa_zmiennej = wartosc
```

Bez deklarowania typu z góry — Python sam rozpozna, czy to liczba, tekst, czy coś innego.


````python
# Przykład — kilka zmiennych różnych typów
imie = "Józef"
liczba_uczniow = 24
srednia_ocen = 4.35
czy_zaliczono = True

print(imie, liczba_uczniow, srednia_ocen, czy_zaliczono)
print(type(imie), type(liczba_uczniow), type(srednia_ocen), type(czy_zaliczono))
````

## 3. Liczby całkowite (`int`) vs zmiennoprzecinkowe (`float`)

- `int` — liczby całkowite, np. `5`, `-3`, `1000`
- `float` — liczby z częścią dziesiętną, np. `5.0`, `19.99`, `-0.5`

Uwaga: `5` to `int`, ale `5.0` to już `float` — sama obecność kropki decyduje o typie.


````python
print(type(5))      # int
print(type(5.0))     # float
print(type(19.99))   # float
````
